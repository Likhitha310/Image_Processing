```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFilter,ImageOps

```

```python
# Create a synthetic image with basic shapes
def create_synthetic_image():
    width, height = 300, 300
    image = Image.new("RGB", (width, height), color="white")
    draw = ImageDraw.Draw(image)

    # Draw a rectangle
    draw.rectangle([50, 50, 200, 200], fill="blue")

    # Draw a circle
    draw.ellipse([220, 50, 270, 100], fill="red")

    # Draw a line
    draw.line([(10, 10), (290, 290)], fill="green", width=3)

    return image
```

```python
# Display the synthetic image
synthetic_image = create_synthetic_image()
plt.imshow(synthetic_image)
plt.title("Synthetic Image")
plt.axis('off')
plt.show()
```

```python
# Convert to grayscale
grayscale_image = synthetic_image.convert("L")
plt.imshow(grayscale_image, cmap="gray")
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
# Edge detection
edge_image = grayscale_image.filter(ImageFilter.FIND_EDGES)
plt.imshow(edge_image, cmap="gray")
plt.title("Edge Detection")
plt.axis('off')
plt.show()
```

```python
# Invert colors
inverted_image = ImageOps.invert(grayscale_image)
plt.imshow(inverted_image, cmap="gray")
plt.title("Inverted Image")
plt.axis('off')
plt.show()
```

```python
# Blur image
blurred_image = synthetic_image.filter(ImageFilter.BLUR)
plt.imshow(blurred_image)
plt.title("Blurred Image")
plt.axis('off')
plt.show()
```

```python
# Rotate image
rotated_image = synthetic_image.rotate(45)
plt.imshow(rotated_image)
plt.title("Rotated Image")
plt.axis('off')
plt.show()
```

```python
# Crop image
cropped_image = synthetic_image.crop((50, 50, 200, 200))
plt.imshow(cropped_image)
plt.title("Cropped Image")
plt.axis('off')
plt.show()
```

```python
# Resize image
resized_image = synthetic_image.resize((150, 150))
plt.imshow(resized_image)
plt.title("Resized Image")
plt.axis('off')
plt.show()
```

```python
# Histogram equalization
equalized_image = ImageOps.equalize(grayscale_image)
plt.imshow(equalized_image, cmap="gray")
plt.title("Histogram Equalization")
plt.axis('off')
plt.show()
```

```python
# Adding text to the image
draw = ImageDraw.Draw(synthetic_image)
draw.text((10, 250), "Hello, Image Processing!", fill="black")
plt.imshow(synthetic_image)
plt.title("Text Added to Image")
plt.axis('off')
plt.show()
```
